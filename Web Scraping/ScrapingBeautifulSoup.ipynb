```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import urllib.request,sys,time
```

```python
pagesToGet= 1

upperframe=[]  
for page in range(1,pagesToGet+50):
    print('processing page :', page)
    url = 'https://1000projects.org/projects/cse-project-reports/page/'+str(page)
    print(url)
    page=requests.get(url)
    time.sleep(2)
    soup=BeautifulSoup(page.text,'html.parser')    
    frame=[]
    paragraphs = soup.find_all("article")
    filename="Projecttrial.csv"
    f=open(filename,"w", encoding = 'utf-8')
    headers="ProjectID,Name,Statement,Link,Overview\n"
    f.write(headers)
    i=1
    for j in paragraphs:
        ProjectID = str(i)
        i+=1
        Name = j.find("h2",attrs={'class':'entry-title'}).text.strip()
        Name = Name.replace(",","^")
        Statement = j.find("div",attrs={'class':'entry-content'}).text.strip()
        Statement = Statement.replace(",","^")
        Link=j.find('a')['href'].strip()
        Overview = j.find_all('p')[0].text.strip()
        Overview = Overview.replace(",","^")
        frame.append((ProjectID,Name,Statement,Link,Overview))
        f.write(ProjectID+","+Name+","+Statement+","+Link+","+Overview+","+"\n")
    upperframe.extend(frame)
f.close()
data=pd.DataFrame(upperframe, columns=['ProjectID,','Name','Statement','Link','Overview'])
print(data)
```
